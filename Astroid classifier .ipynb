```python
def outlier_treatment(col):
    ''' calculating outlier indices and replacing them with NA  '''
    #Extract quartiles
    q75, q25 = np.percentile(df_1[col], [75 ,25])
    print(q75,q25)
    #Calculate IQR
    iqr = q75 - q25
    #Calculate inner and outer fence
    minimum = q25 - (iqr*1.5)
    maximum = q75 + (iqr*1.5)
    print(minimum,maximum)
    #Replace with NA
    df_1.loc[df_1[col] < minimum,col] = np.nan
    df_1.loc[df_1[col] > maximum,col] = np.nan
```

```python
outlier_treatment('Absolute Magnitude')
```
